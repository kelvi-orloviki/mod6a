```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import PyPDF2
%matplotlib inline
```

```python
sinasc_ro_2019 = pd.read_csv('../b1abb33c4f.csv')
sinasc_ro_2019.head()
```

```python
sinasc_ro_2019.columns
```

```python
with open ('../Profissão Cientista de Dados M6 - SINASC.pdf', 'rb') as arquivo_pdf_sinasc:
    sinasc_pdf = PyPDF2.PdfReader(arquivo_pdf_sinasc)
    arquivo_completo = ""
    for numero_pagina in range(len(sinasc_pdf.pages)):
        pagina = sinasc_pdf.pages[numero_pagina]
        arquivo_completo += pagina.extract_text()

print(arquivo_completo)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5', y='CONSULTAS', data=sinasc_ro_2019)
plt.ylabel('Valor Médio por Consulta')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.6, hspace=1)

sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc_ro_2019)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc_ro_2019)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc_ro_2019)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc_ro_2019)

plt.setp([ax.get_xticklabels() for ax in axes.flat], rotation=90)

plt.show()
```
